The above result is in string format but we need a list format to extract genre using function. For that we will be using ast library to do so.


```python
import ast
def converter(obj):
    list =[]
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(converter)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(converter)
```

```python
def converter2(obj):
    list =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            list.append(i['name'])
            counter+=1
        else:
            break
    return list
```

```python
movies['cast'] = movies['cast'].apply(converter2)
```

```python
def converter3(obj):
    list =[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew'] = movies['crew'].apply(converter3)
```

```python
# applying lamda function to make a list instead of a string
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# removing spaces between single names/words this will help us in creating tags 
movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# creating tag features
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['id','title','tags']]
new_df
```

```python
# converting tags to string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
# converting all the tags to lower cases
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
# renaming id to movie_id
new_df.rename(columns = {'id':'movie_id'}, inplace = True)
```

```python
# installing libraries to perform stemming 
!pip install nltk
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return  " ".join(y)
```

```python
# applying stemming on our dataframe
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# max features to select the no of frequently occ. words and stop_words
cv = CountVectorizer(max_features = 5000 ,stop_words = 'english')
# since CountVectorizer will return a scipy sparse matrix we will convert it to a numpy array 
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
# to calculate similarity between a vector(movie)
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
# to get the top 5 recommendation
sorted(list(enumerate(similarity[0])),reverse =True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse =True,key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    return 
```

```python
recommend('Avatar')
```
